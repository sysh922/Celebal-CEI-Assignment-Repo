```python
df["discount"] = df["discount"].fillna(df["discount"].mean())
#as discount value is in float and less number of null value so we can replace null value to mean of data.
```

```python
df["what_customers_said"] = df["what_customers_said"].fillna("No Review")
#filled missing values with placeholder text (No Review) instead of deleting rows.
```

```python
df["seller_name"] = df["seller_name"].fillna("Unknown Seller")
#requried or important data and less number of null value.
```

```python
df = df.drop(columns=["videos"])
#781 out of 1000 are missing values. Not useful for your assignment because mostly empty.
```

```python
df["seller_information"] = df["seller_name"].fillna("Unknown")
# Requried data and less number of missing values so replace with "Unknown".
```

```python
df = df.drop(columns=["variations"])
#So many missing value(562).
```

```python
df["final_price"] = df["final_price"].str.replace('"', '')
#for doing filtering on the basis of "final_price", first we remove double quotes(" "), currency(‚āĻ) symbol and comma.
```

```python
df["final_price"] = df["final_price"].str.replace("‚āĻ", "")
df["final_price"] = df["final_price"].str.replace(",", "")
```

```python
df["final_price"] = pd.to_numeric(df["final_price"], errors="coerce")
#converting column "final-price" from string to number.
```

```python
# As no duplicate data is present so now create derived column as "total_amount = price * quantity".
# Since, there is no "quantity" column avilable so we create one.

df["quantity"] = 1
```

```python
#Derived Column:
df["total_amount"] = df["final_price"] * df["quantity"]
```
